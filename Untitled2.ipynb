```python
import pandas as pd

# Load Excel file
file_path = "/content/GlobalSuperstore.xlsx"
df = pd.read_excel(file_path)
```

```python
# Clean data
df["Sales"] = pd.to_numeric(df["Sales"], errors="coerce")
df["Quantity"] = pd.to_numeric(df["Quantity"], errors="coerce")
df.dropna(subset=["Product Name", "Sales", "Quantity", "Customer ID", "Category", "Order ID"], inplace=True)
```

```python
# Map Phase
product_quantity_kv = df[["Product Name", "Quantity"]]
df["Revenue"] = df["Sales"] * df["Quantity"]
category_revenue_kv = df[["Category", "Revenue"]]
customer_frequency_kv = df.groupby("Customer ID")["Order ID"].nunique().reset_index(name="Frequency")
```

```python
# Reduce Phase
top_5_products = product_quantity_kv.groupby("Product Name").sum().sort_values("Quantity", ascending=False).head(5).reset_index()
revenue_by_category = category_revenue_kv.groupby("Category").sum().reset_index()
top_5_customers = customer_frequency_kv.sort_values("Frequency", ascending=False).head(5)
```

```python
# Save to Excel
with pd.ExcelWriter("MapReduce_GlobalStore.xlsx", engine="xlsxwriter") as writer:
    product_quantity_kv.to_excel(writer, sheet_name="Product_Quantity", index=False)
    category_revenue_kv.to_excel(writer, sheet_name="Category_Revenue", index=False)
    customer_frequency_kv.to_excel(writer, sheet_name="Customer_Frequency", index=False)
    top_5_products.to_excel(writer, sheet_name="Top_5_Products", index=False)
    revenue_by_category.to_excel(writer, sheet_name="Revenue_by_Category", index=False)
    top_5_customers.to_excel(writer, sheet_name="Top_5_Customers", index=False)
```

```python
# 1. Product → Quantity
product_quantity_kv = df[["Product Name", "Quantity"]]

# 2. Category → Revenue (Quantity × Sales)
df["Revenue"] = df["Quantity"] * df["Sales"]
category_revenue_kv = df[["Category", "Revenue"]]

# 3. CustomerID → Frequency (Number of Unique Orders)
customer_frequency_kv = df.groupby("Customer ID")["Order ID"].nunique().reset_index()
customer_frequency_kv.columns = ["Customer ID", "Frequency"]
```

```python
# 1. Top 5 Selling Products
top_5_products = (
    product_quantity_kv.groupby("Product Name")
    .sum()
    .sort_values("Quantity", ascending=False)
    .head(5)
    .reset_index()
)
top_5_products
```

```python
# 2. Revenue by Category
revenue_by_category = (
    category_revenue_kv.groupby("Category")
    .sum()
    .sort_values("Revenue", ascending=False)
    .reset_index()
)
revenue_by_category
```

```python
# 3. Most Frequent Buyers
top_5_customers = (
    customer_frequency_kv.sort_values("Frequency", ascending=False)
    .head(5)
    .reset_index(drop=True)
)
top_5_customers
```
